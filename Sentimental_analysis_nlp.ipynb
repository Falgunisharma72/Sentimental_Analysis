```python
import re
# First Remove all the numbers

def remove_numbers(text):
  return ' '.join([i for i in str(text).split() if not i.isdigit()])

combined_df['WithoutDigit'] = combined_df['message to examine'].apply(lambda x: remove_numbers(x))
```

```python
# Lowercasing all the tweets

combined_df['WithoutDigit'] = combined_df['WithoutDigit'].str.lower()
```

```python
# Removal of Weblinks

def remove_weblinks(text):
  return re.sub(r"http\S+", "", text)

combined_df['WithoutWeblinks'] = combined_df['WithoutDigit'].apply(lambda x: remove_weblinks(x))
```

```python
#Removing twitter mentions
def remove_twitter(text):
  return re.sub('@[\w]+','',text)
combined_df['WithoutMentions'] = combined_df['WithoutWeblinks'].apply(lambda x:remove_twitter(x))
```

```python
import string
def remove_punctuation(text):
    return text.translate((str.maketrans('','', string.punctuation)))
combined_df['WithoutPunctuations'] = combined_df['WithoutMentions'].apply(lambda x:remove_punctuation(x))
```

```python
#Lemmatization
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

wordnet_map = {"n": wordnet.NOUN, "v": wordnet.VERB, "j": wordnet.ADJ, "r": wordnet.ADV}

def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.VERB)) for word, pos in pos_tagged_text])

combined_df["text_lemmatized"] = combined_df['WithoutRareWords'].apply(lambda text: lemmatize_words(text))
```

```python
short_words = {
"aint": "am not",
"arent": "are not",
"cant": "cannot",
"'cause": "because",
"couldve": "could have",
"couldnt": "could not",
"didnt": "did not",
"doesnt": "does not",
"dont": "do not",
"hadnt": "had not",
"hasnt": "has not",
"havent": "have not",
"im": "I am",
"em": "them",
"ive": "I have",
"isnt": "is not",
"lets": "let us",
"theyre": "they are",
"theyve": "they have",
"wasnt": "was not",
"well": "we will",
"were": "we are",
"werent": "were not",
"you're": "you are",
"you've": "you have"
}
```

```python
def replace_short_words(text):
  for word in text.split():
    if word in short_words:
      text = text.replace(word, short_words[word])

  return text
combined_df["Shortened_text"] = combined_df['text_lemmatized'].apply(lambda text:replace_short_words(text))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
import tensorflow_hub as hub
```

```python
# Load Pretrained Word2Vec

embed = hub.load("https://tfhub.dev/google/Wiki-words-250/2")
```

```python
def get_max_length(df):
  ## get max token counts from train data,
  ## so we use this number as fixed length input to RNN cell

  max_length = 0
  for row in new_table['Tweets']:
    if len(row.split(" ")) > max_length:
      max_length = len(row.split(" "))

  return max_length
```

```python
get_max_length(new_table['Tweets'])
```

```python
def get_word2vec_enc(tweets):
  ## get word2vec value for each word in sentence
  # concatenate word in numpy array, so we can use it as RNN input

  encoded_tweets = []
  for tweet in tweets:
    tokens = tweet.split(" ")
    word2vec_embedding = embed(tokens)
    encoded_tweets.append(word2vec_embedding)
  return encoded_tweets
```

```python
def get_padded_encoded_tweets(encoded_tweets):
  # for short sentences, we prepend zero padding so all input to RNN
  # has same length

  padded_tweets_encoding = []
  for enc_tweet in encoded_tweets:
    zero_padding_cnt = max_length - enc_tweet.shape[0]
    pad = np.zeros((1, 250))
    for i in range(zero_padding_cnt):
      enc_tweet = np.concatenate((pad, enc_tweet), axis = 0)
    padded_tweets_encoding.append(enc_tweet)
  return padded_tweets_encoding
```

```python
def sentiment_encode(sentiment):
    if sentiment == 0:
        return [0,1]
    else:
        return [1,0]
```

```python
def preprocess(df):
  # encode text value to numeric value

  tweets = df['Tweets'].tolist()

  encoded_tweets = get_word2vec_enc(tweets)
  padded_encoded_tweets = get_padded_encoded_tweets(encoded_tweets)

  #encoded sentiment
  sentiments = df['Labels'].tolist()
  encoded_sentiment = [sentiment_encode(sentiment) for sentiment in sentiments]

  X = np.array(padded_encoded_tweets)
  Y = np.array(encoded_sentiment)

  return X, Y
```

```python
# Preprocess

max_length = get_max_length(new_table)
max_length
```

```python
tdf = new_table.sample(frac = 1)
train = tdf[:3702]
test = tdf[3702:]
```

```python
train.shape, test.shape
```

```python
train_X, train_Y = preprocess(train)
test_X, test_Y = preprocess(test)
```

```python
# LSTM

model = Sequential()
model.add(LSTM(32))
model.add(Dense(2, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
#model.fit(train_X, train_Y, epochs = 10)#with epoch 10 the accuracy came out to be 0.9868
model.fit(train_X, train_Y, epochs = 15)
```

```python
model.summary()
```

```python
score, acc = model.evaluate(test_X, test_Y, verbose = 2)
print("Test Score:", score)
print("Test Accuracy:", acc)
```

```python
# Confusion Matrix

y_pred = model.predict(test_X)
```
